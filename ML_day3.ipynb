Ordinal Encoding ---it can be perform in 2 ways sklearn and map function


```python
import pandas as pd
```

```python
dfs=pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","xl","m"]})
dfs.head(3)
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(dfs[["Size"]])
```

```python
dfs["size_en"]=oe.transform(dfs[["Size"]])
dfs
```

By map function

```python
ord_data1={"s":0,"m":1,"l":2,"xl":3}
```

```python
dfs["Size_en_map"]=dfs["Size"].map(ord_data1)
```

```python
dfs
```
