```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset.drop(columns=["User ID","Gender","EstimatedSalary"],inplace=True)
dataset.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=dataset)
plt.show()
```

```python
x=dataset[["Age"]] #2 dimensional
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict([[47]])
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.lineplot(x="Age",y=lr.predict(x),data=dataset,color="red")
plt.show()
```

```python
datasets=pd.read_csv("placement (1).csv")
datasets.rename(columns={'resume_score':'score'},inplace=True)
datasets.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="score",data=datasets,hue="placed")
plt.legend(loc=1)
plt.show()
```

```python
x=datasets.iloc[:,:-1]
y=datasets["placed"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
la=LogisticRegression()
la.fit(x_train,y_train)
```

```python
la.score(x_test,y_test) #1.0 means 100 percent
```

```python
la.predict([[8.14,6.52]])

```

```python
la.coef_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=la)
plt.show()
```

Logistic Regression (practical) (Binary Classification) (Polynomial input)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_circles

n_samples = 1000

x, y = make_circles(n_samples,
                    noise=0.03,
                    random_state=42)

len(x),len(y)
```

```python
print(f"First five samples of X: {x[:5]}")
print(f"First five targets of y: {y[:5]}")
```

```python
import pandas as pd

circles = pd.DataFrame({"X0": x[:, 0],
                       "X1": x[:, 1],
                       "y": y})
circles.head(10)
```

```python
sns.scatterplot(x="X0",y="X1",data=circles,hue="y")
```

```python
p=circles.iloc[:,:-1]
q=circles["y"]
```

```python
from sklearn.model_selection import train_test_split
p_train,p_test,q_train,q_test=train_test_split(p,q,test_size=0.2,random_state=44)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo=LogisticRegression()
lo.fit(p_train,q_train)
```

```python
lo.score(p_test,q_test)
```
