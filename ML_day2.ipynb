```python
datas=pd.read_csv("Vehicle Price.csv")
print(datas)
```

```python
datas.isnull().sum()
```

```python
datas.select_dtypes(include="float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
ar=si.fit_transform(datas[['price', 'cylinders', 'mileage', 'doors']])
```

```python
new_dataset=pd.DataFrame(ar,columns=datas.select_dtypes(include="float64").columns )
```
